```python
import re
def fix_phone(phone_num):
    if re.match(r'.63',phone_num):
        return re.sub(r'(.63)\s',r'0',phone_num)
    elif re.match(r'9',phone_num):
        return re.sub(r'([0-9]{3}\s[0-9]{3}\s[0-9]{4})',r'0\1',phone_num)
    else:
        return phone_num
```

```python
vfunc = np.vectorize(fix_phone)
recweek['phonenumber'] = vfunc(recweek['phonenumber'])
```

```python
# write code here
phone_df = merged_df.loc[:,["gender",'phonenumber']]
```

```python
def carrier_identifier(phonenumber):
    if re.match('0917',phonenumber) or re.match('0927',phonenumber):
        return 'Globe'
    elif re.match('0918',phonenumber):
        return 'Smart'
```

```python
vec_carrier = np.vectorize(carrier_identifier)
phone_df['phonenumber'] = vec_carrier(phone_df['phonenumber'])
```

```python
phone_df
```

```python
grouped_carrier = phone_df.groupby('phonenumber')
grouped_carrier = grouped_carrier.count()
```
